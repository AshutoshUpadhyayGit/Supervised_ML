```python
# Part I

# create design matrix X and target vector y
#  All input features
# X = np.array(vectorized_text_df.iloc[:, :-1]) # select all colums except the last column 

# X = np.array(vectorized_text_df.iloc[:,1]) # select column at index 1 because thats our vector form of Text
# # Target column
# y = np.array(vectorized_text_df['Score']) # showing you two ways of indexing a pandas df

X = np.array(vectorized_text_df['tfidf_sent_vectors'].to_list())
y = np.array(vectorized_text_df['Score'].to_list())
```

### Tuning Hyperparamater C: Simple Cross Validation ###
####Tuning Hyperparamater C to find optimal value of C to avoid Overfitting / Underfitting of model####

```python
# split the data set into train and test
X_1, X_test, y_1, y_test = model_selection.train_test_split(X, y, test_size=0.3, random_state=0)

# split the train data set into cross validation train and cross validation test
X_tr, X_cv, y_tr, y_cv = model_selection.train_test_split(X_1, y_1, test_size=0.3)


for i in [10**-4, 10**-2, 10**0,10,10**2, 10**4]:
    model = LogisticRegression(penalty='l2',solver='liblinear',multi_class ='ovr',C=i )

    # fitting the model on train dataset
    model.fit(X_tr, y_tr)

    # predict the response(predicd = predicted y_cv) on the 
    # crossvalidation dataset(X_cv) on the above trained model on training data with K = i
    pred = model.predict(X_cv)

    # evaluate CV accuracy on K = i
    acc = accuracy_score(y_cv, pred, normalize=True) * float(100)
    print('\nCV accuracy for C = ' + str(i) + ' is '+str(acc) )


    # F1 Score of each class
    # Note: labels are class labels required for multiclass classification
    # F1 score is provided for each class showing how well model is doing on predcting each class
    # Higher F1 score the better
    f1_score_model = f1_score(y_cv, pred,labels=[0,1,2],average=None)     

    # print('\nCV accuracy for k = %d is %d%%' % (i, acc))
    print('\nCV f1_score_model for k = ' + str(i) + " is" + str(f1_score_model))


```

#### Retraining with the optimal value of K received using simple cross validation and calculating test accuracy ####

```python
# Retrain the model with tuned hyperparmeter Coptimal (Coptimal = C where accuracy is max in CV) 
model = LogisticRegression(penalty='l2',solver='liblinear',multi_class ='ovr',C=100 )

# fitting the model on train dataset
model.fit(X_tr, y_tr)

# predict the response(pred = predicted y) on the 
# test dataset(X_test) on the above trained model with optimal C
pred = model.predict(X_test)
print('Predicted output on test data:')
print(pred)

# evaluate CV accuracy on K = i
acc = accuracy_score(y_test, pred, normalize=True) * float(100)
print('\nTest accuracy for C = 100 : ',str(acc) )

#  Predicted class from KNN

# Get text corresponding to vectors
# actual_text = vectorized_text_df.loc[vectorized_text_df['tfidf_sent_vectors'].isin(list(X_test))]['Text']
# print(actual_text)

model_predicted_on_test_data = pd.DataFrame()
model_predicted_on_test_data['Text'] = list(X_test)
model_predicted_on_test_data['Actual Class'] = y_test
model_predicted_on_test_data['Predicted Class'] = pred

model_predicted_on_test_data.to_csv(r"Logistic_Reg_Predicted_Class.csv",index=False)
model_predicted_on_test_data.head(5)
```

### Hypertuning using RandomizedSearchCV and 5Fold CV

```python
#Using RandomizedSearchCV
from scipy.stats import uniform
distributions = dict(C=np.arange(10**-4 , 10**4))

model = RandomizedSearchCV(LogisticRegression(multi_class='ovr',penalty='l2'), distributions, scoring = 'accuracy', cv=5)
# If using F1 score as evaluation metric ; scoring ='f1_macro'
model.fit(X_tr, y_tr)

print(model.best_estimator_)
print(model.score(X_test, y_test))
```

### 10 fold Cross vaildation ###

```python


# creating odd list of K for KNN
myList = [10**-4, 10**-2, 10**0,10,10**2, 10**4]

# empty dict that will hold cv scores
cv_scores = {}

# perform CV = 10-fold cross validation
for k in myList:
    knn = LogisticRegression(multi_class='ovr',penalty='l2',C=k)
    scores = cross_val_score(knn, X_tr, y_tr, cv=10, scoring='accuracy')
    cv_scores[k] = scores.mean()

print(cv_scores)
print("Optimal C value:")

optimal_k = max(cv_scores, key=cv_scores.get)
print(optimal_k)
    
```
